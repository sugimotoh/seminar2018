```
from bs4 import BeautifulSoup
import requests
import time
```

```
txt_all = []
for i in range(1,20):
    if i == 1:
        target_url = 'ここにURL'
    else:
        target_url = 'ここにURL?page='+str(i)
        
    r = requests.get(target_url)
    soup = BeautifulSoup(r.text, "html.parser")
    txt = soup.find_all(class_="os1-symbol-numerical mg10l")
    txt_all.append(txt)
    time.sleep(3)
```

```
str_txt_all = str(txt_all)
str_txt_all2 = str_txt_all.replace('<span class="os1-symbol-numerical mg10l">', "").replace("</span>", "").replace("[","").replace("]","")
str_txt_all3 = str_txt_all2.replace(',', "")
data01=[float(i) for i in str_txt_all3.split(' ')]
data01
```

```
from statistics import mean, median, variance, stdev

length = len(data01)                         # サンプルサイズ, リストの要素数(長さ)
print('サンプルサイズ',length)

print('合計',"{0:.4f}".format(sum(data01)))  # データの合計

ave = sum(data01)/len(data01)                # データの平均
print('平均',"{0:.4f}".format(ave))

variance = 0                                 # データの分散
for member in data01:
    variance = variance+(member-ave)**2      
variance = variance/length                   
print('分散',"{0:.4f}".format(variance))

# print('分散',"{0:.4f}".format(variance))   # 統計関数による検算
```
