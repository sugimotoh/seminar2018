```
import pandas as pd
df_goods = pd.read_excel("/content/drive/MyDrive/data_open/商品.xls")
df_sales = pd.read_excel("/content/drive/MyDrive/data_open/売上.xls")
df_custm = pd.read_excel("/content/drive/MyDrive/data_open/顧客.xls")
```

```
df_goods
df_sales
# df_custm
```

```
df_goods_sales = pd.merge(df_goods,df_sales,on='商品ID',how='inner')
df_goods_sales
```

```
df_goods_sales_custm = pd.merge(df_goods_sales,df_custm,on='顧客ID',how='inner')
```
