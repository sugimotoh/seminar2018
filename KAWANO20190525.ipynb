```
words = ""
for token in tokens:
    if token.part_of_speech.split(',')[0] in ['名詞', '動詞', '形容詞', '副詞']:
        words = words + " " + token.base_form
```

```
from wordcloud import WordCloud
fpath = '/usr/share/fonts/opentype/ipafont-gothic/ipagp.ttf'
wordcloud = WordCloud(background_color="white"
                     ,width=800
                     ,height=500).generate(words)
```

```
stop_words = [u'する', u'ある', u'こと', u'いる', u'れる'
             ,u'よう', u'やる', u'もの', u'lace', u'つく', u'Ｔステ']

wordcloud = WordCloud(background_color="white"
                      ,font_path=fpath
                      ,width=800
                      ,height=500
                      ,stopwords=set(stop_words)).generate(words)
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(15,9))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
