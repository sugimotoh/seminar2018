**1-4. 棒グラフの作成（seabornの利用）**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ndarrayに変換
data = np.array(data_MM01['HINCOME'])

sns.set()
sns.set_style('whitegrid')
sns.set_palette('pastel') 

# ヒストグラム
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(data, bins=10, histtype='barstacked', ec='white',alpha=1.0)
plt.show()
```

**1-5. 棒グラフの作成（plotyの利用※推奨）**

```python
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
pio.templates["draft"] = go.layout.Template(
    layout_annotations=[
        dict(
            name="draft watermark",
            text="fukuoka univ",
            textangle=-30,
            opacity=0.1,
            font=dict(color="black", size=100),
            xref="paper",
            yref="paper",
            x=0.5,
            y=0.5,
            showarrow=False,
        )
    ]
)
pio.templates.default = "plotly_white+draft"
fig = px.histogram(data_MM01, x="HINCOME",  histnorm='probability density',color_discrete_sequence=['indianred'])
fig.update_layout(bargap=0.1,title='所得分布')
fig.show()
```

```python
fig = px.histogram(data_MM01, x="Q6S1",  nbins=11,color_discrete_sequence=["green"])
fig.update_layout(bargap=0.1,title='伊藤園 お～いお茶　0点:非常に薦めたい,11点:全く薦めたくない')
fig.show()
```
